```python
# Convert the 'date' column to datetime objects
df['date'] = pd.to_datetime(df['date'])

# Extract numerical features from the date
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day

# Now calculate the correlation, dropping the original 'date' column
df.drop('date', axis=1).corr()['price'].sort_values()
```

```python
features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]
```

```python
X = df[features]
y = df['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=1)
model.fit(X_train, y_train)
r2 = r2_score(y_test, model.predict(X_test))
print(f'R^2: {r2:.2f}')
```

```python
model = Ridge(alpha=0.1)
X = df[features]
y = df['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=1)
model.fit(X_train, y_train)
r2 = r2_score(y_test, model.predict(X_test))
print(f'R^2: {r2:.2f}')
```

```python
poly_features = PolynomialFeatures(degree=2)
model = Ridge(alpha=0.1)
pipeline = make_pipeline(poly_features, model)
X = df[features]
y = df['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=1)
pipeline.fit(X_train, y_train)
r2 = r2_score(y_test, pipeline.predict(X_test))
print(f'R^2: {r2:.2f}')
```
